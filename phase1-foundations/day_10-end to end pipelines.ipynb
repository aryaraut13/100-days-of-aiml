```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

print("=" * 55)
print("DATA PIPELINE v1.0 — Heart Disease Dataset")
print("Week 1 wrap-up: NumPy → Pandas → EDA → Clean → Engineer")
print("=" * 55)

url = "https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv"
df  = pd.read_csv(url)
print(f"\n[LOAD]     {df.shape[0]} rows × {df.shape[1]} features")
print(df.head(3).to_string())

print(f"\n[EXPLORE]  Nulls: {df.isnull().sum().sum()} | "
      f"Duplicates: {df.duplicated().sum()}")
print(f"           Target balance:\n"
      f"{df['target'].value_counts(normalize=True).round(2).to_string()}")


df.drop_duplicates(inplace=True)
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
print(f"\n[CLEAN]    Rows after dedup: {df.shape[0]} | "
      f"Nulls remaining: {df.isnull().sum().sum()}")


df["age_group"]    = pd.cut(df["age"],
                             bins=[0, 40, 55, 70, 100],
                             labels=["Young","Mid","Senior","Elderly"])
df["high_chol"]    = (df["chol"] > 240).astype(int)
df["hr_age_ratio"] = (df["thalach"] / df["age"]).round(3)
print(f"\n[FEATURES] Created: age_group | high_chol | hr_age_ratio")


scaler       = StandardScaler()
cols_to_scale = ["age","trestbps","chol","thalach","oldpeak"]
df[cols_to_scale] = scaler.fit_transform(df[cols_to_scale])
print(f"[SCALE]    Standardized: {cols_to_scale}")


df.to_csv("heart_processed.csv", index=False)

print(f"\n[DONE]     Final shape: {df.shape}")
print(f"           Saved → heart_processed.csv ✓")
print(f"           Ready for modeling.")
print(f"\nAll features: {list(df.columns)}")
print("\n" + "=" * 55)
print("Week 1 complete. 10 days. 10 commits. 0 models.")
print("Next week: we train on this.")
print("=" * 55)
```
