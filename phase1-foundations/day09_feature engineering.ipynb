```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Fare"].fillna(df["Fare"].median(), inplace=True)

print("=" * 50)
print("FEATURE ENGINEERING — Creating signal from noise")
print("=" * 50)


df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"]    = (df["FamilySize"] == 1).astype(int)
df["Title"]      = df["Name"].str.extract(r" ([A-Za-z]+)\.")
df["AgeBin"]     = pd.cut(df["Age"],
                           bins=[0, 12, 18, 60, 100],
                           labels=["Child","Teen","Adult","Senior"])
df["FareBin"]    = pd.cut(df["Fare"],
                           bins=4,
                           labels=["Low","Mid","High","VHigh"])

print("\nTitles extracted from raw names:")
print(df["Title"].value_counts().head(8))

print("\nSample — new features:")
print(df[["Name","Title","FamilySize","IsAlone","AgeBin"]].head(6).to_string())


scaler_std = StandardScaler()
scaler_mm  = MinMaxScaler()

df["Age_std"]  = scaler_std.fit_transform(df[["Age"]])
df["Fare_mm"]  = scaler_mm.fit_transform(df[["Fare"]])

print("\n" + "─" * 50)
print("SCALING COMPARISON:")
print("─" * 50)
print(f"Age original  → mean: {df['Age'].mean():.1f}  | std: {df['Age'].std():.1f}")
print(f"Age scaled    → mean: {df['Age_std'].mean():.4f} | std: {df['Age_std'].std():.4f}")
print(f"\nFare original → min: {df['Fare'].min():.1f}  | max: {df['Fare'].max():.1f}")
print(f"Fare MinMax   → min: {df['Fare_mm'].min():.1f}   | max: {df['Fare_mm'].max():.1f}")

print(f"\nFeatures created: {['FamilySize','IsAlone','Title','AgeBin','FareBin']}")
print(f"Total columns now: {df.shape[1]}")
```
