```python
import numpy as np

print("=" * 50)
print("LINEAR ALGEBRA — The engine of every AI model")
print("=" * 50)

v1 = np.array([3, 4])
v2 = np.array([1, 2])

print(f"\nDot product:       {np.dot(v1, v2)}")
print(f"Magnitude of v1:   {np.linalg.norm(v1):.2f}")
print(f"Cosine similarity: {np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)):.4f}")
print("↑ This is how every search in a RAG pipeline works")

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("\nA @ B:")
print(A @ B)
print("\nA.T:")
print(A.T)

print("\n" + "─"*50)
print("THE FORWARD PASS OF A NEURAL NETWORK:")
print("─"*50)

data    = np.random.randn(100, 5)
weights = np.random.randn(5, 1)
output  = data @ weights

print(f"Input:   {data.shape}")
print(f"Weights: {weights.shape}")
print(f"Output:  {output.shape}")
print("\nThat's it. Stack 50 of these with activations = GPT.")
```
