```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = 2 * np.random.rand(100)
y = 4 + 3 * X + np.random.randn(100)

theta_0 = 0.0
theta_1 = 0.0
lr = 0.1
n = len(X)
losses = []

for epoch in range(100):
    y_pred = theta_0 + theta_1 * X
    loss = np.mean((y_pred - y) ** 2)
    losses.append(loss)
    d0 = (2/n) * np.sum(y_pred - y)
    d1 = (2/n) * np.sum((y_pred - y) * X)
    theta_0 -= lr * d0
    theta_1 -= lr * d1

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
fig.suptitle("Gradient Descent — Learning from scratch", fontsize=13)

axes[0].plot(losses, color="#5865F2", lw=2)
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Loss (MSE)")
axes[0].set_title("Loss Curve")

axes[1].scatter(X, y, alpha=0.5, color="#57F287", label="Data")
axes[1].plot(X, theta_0 + theta_1 * X, color="#ED4245", lw=2,
             label=f"θ₀={theta_0:.2f}, θ₁={theta_1:.2f}")
axes[1].legend()
axes[1].set_title("Final Fit")

plt.tight_layout()
plt.savefig("day11_gradient_descent.png", dpi=150)
plt.show()

print(f"Final θ₀: {theta_0:.4f}  (true: 4.0)")
print(f"Final θ₁: {theta_1:.4f}  (true: 3.0)")
print(f"Loss: {losses[0]:.2f} → {losses[-1]:.4f}")
```
